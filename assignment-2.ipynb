## USING REGEX
n
A RegEx, or Regular Expression, is a sequence of characters that forms a search patter
. RegEx can be used to check if a string contains the specified search pattern

```python
import regex as re # importing and using regex for  matching and extracting string patterns

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

## TESTTOKENS

```python
testtext = "im feeling a little like a damaged tree and that my roots are a little out of wack"
testtokens = tokenize(testtext) # splitting up a larger body of text into smaller lines, words or even creating words for a non-English language.
 
print(testtokens)
```

## IMPORTING NLTK PACAKAGE
 variety of tasks can be performed using NLTK such as tokenizing, parse tree visualization,

```python
import nltk # importing Natural Language Toolkit for  which helps in tokenization..
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

## STOP WORDS

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

## TOKENS

```python
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

## TEXT PREPROCESSING PIPELINE

```python
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
df['num_tokens'] = df['tokens'].map(len)
df.tail()
# returns last rows
```

## USING COLLECTION AND IMPORTING COUNTER

Counter. A Counter is a data type in the collections module. It is a dict subclass that stores counts for hashable objects.

```python
from collections import Counter # importing counter to count variables

tokens = tokenize("She likes my cats and my cats like my sofa.")

testcount= Counter(tokens)
print(testcount)
```

## TOKENIZATION
Tokenization is a preprocessing technique in natural language processing (NLP).
It breaks down unstructured text data into smaller units called tokens.

```python
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

## MAPPING

```python
counter = Counter() #It is used to count the occurrences of elements in a collection (typically a list or string).
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

## TOKENIZATION

```python
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
freq_df = freq_df.sort_values('freq', ascending=False) # sorting according to frequency
freq_df.head(10)
```

## DESCRIPTION

```python
freq_df.describe() 
#The df.describe() method provides a statistical summary of the numerical columns.
```

## WORDCLOUD
.
Word cloud is one of the most powerful and straightforward visualization methods when it comes to text data
. The size of words are dependent on the occurrence frequency

This below is a wordcloud sample for the dataset emotions.It is a basic wordcloud where we have not removed stopwords nor we added any masking to it.

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Reading text from CSV file
df = pd.read_csv('text.csv')

# Combining text from all rows
all_text = ' '.join(df['text'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Plotting word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## WORDCLOUD-2 MASK IMAGE

This is a masked wordcloud representing the word frequency in the humamn face

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Read text from CSV file
#df = pd.read_csv('text.csv')
#all_text = ' '.join(df['text'])
mask_image = np.array(Image.open("human.jpg")) 

custom_stopwords = set(STOPWORDS)

wordcloud = WordCloud(width=1000, height=800, background_color='white',mask=mask_image, stopwords=custom_stopwords).generate(all_text)

# Plotting word cloud
plt.figure(figsize=(12, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
